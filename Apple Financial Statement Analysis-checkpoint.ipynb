```python
from yahoofinancials import YahooFinancials
```

```python
ticker = 'AAPL'
yf = YahooFinancials(ticker)
income_statement =  yf.get_financial_stmts('annual','income')
```

```python
aapl_is = income_statement['incomeStatementHistory']['AAPL']
```

```python
import pandas as pd 
```

```python
df_list = []

for d in aapl_is:
    df_list.append(pd.DataFrame.from_dict(d, orient= 'index'))

aapl_is =  pd.concat(df_list)
```

```python
aapl_is.head(5)
```

```python
print(aapl_is)
```

```python
import pandas as pd

# Assuming 'aapl_is' is your DataFrame
# Convert the index to datetime if it's not already
aapl_is.index = pd.to_datetime(aapl_is.index)

# Create a new column named 'date' from the index
aapl_is['date'] = aapl_is.index

# Print the DataFrame to see the new column
aapl_is.head(5)
```

```python
aapl_is.info()
```

```python
aapl_is['Year'] = aapl_is['date'].dt.year
aapl_is['Year'].describe()
```

```python
# Export DataFrame to Excel
excel_file_path = 'apple_incomestatement.xlsx'  # Specify the path where you want to save the Excel file
aapl_is.to_excel(excel_file_path, index=False)  # index=False to exclude DataFrame index from Excel
```
